###1.4.2 Самые важные hard и soft скилы

Для определения топ hard навыков объеденим все значения из ячеек `hard_skills_from_description` в один список и посчитаем частоту вхождения навыков в список. Аналогичным образом поступим с soft навыками.  
Для этого создадим функию, которая объеденит навыки в общий список.

```python
def flatten(lst):
    result = []
    for i in lst:
        if isinstance(i, list):
            result.extend(flatten(i))
        else:
            result.append(i)
    return result
```

Выполним фильтрацию данных по категории опыта и применим функцию `flatten(lst)` к соотвествующим столбцам скилов для Аналитиков данных и Бизнес аналитиков.
